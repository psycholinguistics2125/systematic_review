```python
import pandas as pd
```

```python
row_data = pd.read_excel("data/PTSD-Language_full_text_screening.xlsx")
row_data.columns = row_data.iloc[0]
row_data = row_data[1:]
```

```python
select = ["date_year","date_month","date_day","Title","Authors","Abstract","Article URL"]
data = row_data[select]



data.rename(columns={"date_year":"Publication Year",
                     "date_month":"Publication Month",
                     "date_day":"Publication Day",
                     "Title":"Title",
                     "Authors":"Authors",
                     "Abstract":"Abstract",
                     "Article URL":"Source"}, inplace=True)

data['Abstract'] = data['Abstract'].str.replace('\n', ' ')
data.to_csv("data/PTSD-Language_full_text_screening.tsv",sep = "\t" ,index=False)
data.head()
```
